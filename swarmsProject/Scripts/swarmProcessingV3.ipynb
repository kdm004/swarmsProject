```python
import numpy as np
import plotly.graph_objs as go
import plotly.express as px
import scipy.stats as stats

```

```python
# Specify dimensions of array
rows = 34
columns = 10


def Get_Data(filePath):
    caseFile = open(filePath)
    caseData = caseFile.readlines()
    caseFile.close()
    caseData = list(map(float, caseData))
    #print(case1Data)

    caseMatrix = np.zeros((rows,columns))
    #print(case1Matrix)

    caseMatrix[0] = caseData[0:columns]

    for i in range(rows):
        caseMatrix[i] = caseData[i*columns:i*columns+columns]
        
    return caseData, caseMatrix

```

```python
case1Data, case1Matrix = Get_Data('../swarmData/obstacleEnv_fitnesses_case1_inOrder.txt')
case2Data,case2Matrix = Get_Data('../swarmData/obstacleEnv_fitnesses_case2.txt')

numberOfBots1 = [0.9, 1.9, 2.9, 3.9, 4.9, 5.9, 6.9, 7.9, 8.9, 9.9]*34
numberOfBots2 = [1.1, 2.1, 3.1, 4.1, 5.1, 6.1, 7.1, 8.1, 9.1, 10.1]*34


#numberOfBots2 = [1.1,2.1,3.1,4.1,5.1,6.1,7.1,8.1,9.1,10.1]*34

```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=numberOfBots1,y=case1Data,name='sameBody, sameBrain',mode='markers'))
fig.add_trace(go.Scatter(x=numberOfBots2,y=case2Data,name='sameBody, diffBrain',mode='markers'))
fig.update_layout(yaxis_range=[-15,15])
fig.update_xaxes(title='Robot in Swarm')
fig.update_xaxes(nticks=20)
fig.update_yaxes(title='Displacement in x-direction')
fig.update_layout(title=f'Fitness Comparison for 34 Runs of Each Swarm')
fig.update_yaxes(autorange="reversed")

fig.show()
```

```python
# ---------------------------------paired t-test--------------------------------------
```

```python
case1Avgs = []
case2Avgs = []
for i in range(34-1):
    case1Avgs.append(np.mean(case1Matrix[i]))
    case2Avgs.append(np.mean(case2Matrix[i]))
    
# print(case1Avgs[0])
# print(case2Avgs[0])

results = stats.ttest_rel(case1Avgs, case2Avgs) # two sided t-test
print('Results:', results)



# Result of previous t-test: (best controller of 10 runs was used )
# Results: Ttest_relResult(statistic=-1.6489179104848708, pvalue=0.10894908909774198)
```
